## Preprocessing the data

```python
## defining the function to extract the important data in "genres" and "keywords" coloumn
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
```

```python
import ast
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
## defining the function to extract the importatn data in "cast" and "crew" coloumn
def convert2(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3: 
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert2) #using convert2 method, i am trying to extract top 3 casts in the dataframe
```

```python
movies.head()
```

```python
## defining a function to fetch director's name in "credit" coloumn

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# now converting overview data into a list of strings
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```


### Applying transformation inorder to get rid of space between the names which belongs to same entity
#### For eg :  Converting "Sam Worthington" to "SamWorthington" 
Performing this task is necessary because it may occur that there maybe many names containing "Sam" as a first name and the model would get confused and start showing the recommendation of different "Sam" which is incorrect. 

```python
# Using list comprehension to get rid of blank spaces 

movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# creating a new column named "tags" in the movie dataframe and storing all contents of every coloumn in it 

movies['tags'] = movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

```python
movies.head()
```

### Now creating a new dataframe and storing the key columns like "movie_id", "title", "tags"


```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
# now in tags column, converting the list data into string 

new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
# Converting every string to lowercase beacuse it is recommended 
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

### There are many same type of words like "actor" and "actors", both are same but they are representing different entites. 

### Inorder to get of rid those kind of words I am going to using Stemming technique 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
ps.stem('eating')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

### Now performing Text Vectorization using Scikit learn

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

### There are 4806 movies and 5000 most frequent words in the corpose

```python
vectors.shape
```

```python
## Seeing the most common words 

cv.get_feature_names()
```

### I am using cosine similarity to find out how similar the movie is compared to rest of the movies 

#### Cosine similarity ranges from 0 to 1 which means if the value is near 1 then the movie is highly similar and vice versa 

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

#### Now I am defining a function which will return the top five vectors which are similar to the input vector

```python
def recommend(movie):
    #finding the index of a given movie in the dataframe
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
       
```

```python
recommend('Jurassic Park')
```
